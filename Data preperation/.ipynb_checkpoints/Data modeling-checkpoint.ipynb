```python
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 999)
%matplotlib inline
```

```python
df = pd.read_csv('Data/Merged_Verkiezingsuitslag.csv')
```

```python
df.shape
```

```python
df
```

```python
df = df.groupby(['Year','OuderRegioNaam']).sum()
```

```python
df_sub = df.loc[:, df.columns != 'Year' ]
to_remove = []
threshold = 100000
for col in df_sub.columns:
    count_votes = df[col].sum()
    if count_votes < threshold:
        to_remove.append(col)

df = df.drop(to_remove, axis=1)
```

```python
df.to_excel('Data_gefiltered_op_jaar_provencie.xlsx')
```
